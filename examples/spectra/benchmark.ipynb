```python
import os 
import glob

from dotenv import load_dotenv
from datasets import load_dataset
import pandas as pd


load_dotenv() 

# Spender I: 7526202_0 - model numbers
# Spender II: 7527549_0 - model numbers
DATA_ROOT = os.getenv("DATA_ROOT")
```

```python
# Combine all partial results
files = glob.glob("./results_raw_embeddings/*.csv")
df_flower = pd.concat([pd.read_csv(f) for f in files], ignore_index=True)

# Sort for a clean table
df_flower = df_flower.sort_values(['Spender', 'Attribute', 'Split', 'Layers'])
```

```python
# UMAP
```

```python
# 1. Find all CSV files in the results directory
file_pattern = "./results_umap_w_err/*.csv"
all_files = glob.glob(file_pattern)

if not all_files:
    print("No result files found. Check your ./results_umap directory.")
else:
    # 2. Concatenate all files into a single DataFrame
    df_list = [pd.read_csv(f) for f in all_files]
    df_umap = pd.concat(df_list, ignore_index=True)

    # 3. Clean and Sort the table
    # Sorting by Attribute and then Spender makes for the best comparison
    df_umap = df_umap.sort_values(by=["Attribute", "Spender"])
```
